```python
import numpy as np
from scipy.stats import norm
import pandas as pd
```

```python
df=pd.read_csv('price_timeseries_30d_WETH_USDT.csv', sep=',')
```

```python
df
```

```python
q_x = df["token_0"].values[-1]
q_y = df["token_1"].values[-1]
K = 1.2888929e+16
n_days = 29
n_edays = 29
T_x = 1000000
```

```python
K
```

```python
quotes = np.zeros((n_days,2))
quotes[:,0] = df["token_0"].values[:-2]
quotes[:,1] = df["token_1"].values[:-2]
#first column DAI, second column USDC
```

```python
quotes
```

```python
pools = np.zeros((n_days,2))
pools[:,0] = np.sqrt(K * quotes[:,0]/(quotes[:,1]))
pools[:,1] = np.sqrt(K * quotes[:,1]/quotes[:,0])
```

```python
pools
```

```python
# rf=np.ones((n_days,2))
# rf[:,0] = 0.1 / 365
# rf[:,0] = 0.15 / 365
```

```python
gas = df["swap_price"].values[:-2]
```

```python
gas_bar = np.mean(gas) 
```

```python
gas_bar
```

```python
r = np.zeros((n_days-1,2))
r[:,0] = (quotes[1:n_days,0] - quotes[0:n_days - 1,0])/ quotes[0:n_days - 1,0]
r[:,1] = (quotes[1:n_days,1] - quotes[0:n_days - 1,1])/ quotes[0:n_days - 1,1]
```

```python
r
```

```python
mu_x, std_x = norm.fit(r[:,0])
mu_y, std_y = norm.fit(r[:,1])
```

```python
mu_x, std_x
```

```python
e_p = np.zeros((n_edays,2)) 

for i in range(n_edays): 
    e_p[i,0] = 1
    e_p[i,1] = q_y * (np.exp((mu_y + 0.5 * std_y**2)*i)) 
```

```python
e_pools = np.zeros((n_edays,2))
e_pools[:,0] = np.sqrt(K / (e_p[:,1]/e_p[:,0]))
e_pools[:,1] = np.sqrt(K / (e_p[:,0]/e_p[:,1]))
```

```python
e_pools
```

```python
e_p
```

```python
totalEstimatedPrice = np.zeros((n_edays,1)) 
```

```python
def priceIntegral(j,i): 
    #yo = (1 / (3 * K)) * (e_pools[j,0]**3 - (e_pools[j,0] - (T_x / i))**3) * e_p[j,1]
    yo = (K) * ((1/(e_pools[j,0] - (T_x / i))) - (1 / e_pools[j,0])) / e_p[j,1]
    print(yo)
    return (yo)
```

```python
for i in range(1,n_edays): 
    for j in range(i):
        totalEstimatedPrice[i,0] += (priceIntegral(j,i)) + gas_bar 
```

```python
totalEstimatedPrice
```

```python
def main():
    for i in range(1,n_edays): 
        for j in range(i):
            totalEstimatedPrice[i,0] += (priceIntegral(j,i)) + gas_bar 
    

if __name__ == "__main__":
    main()
```
